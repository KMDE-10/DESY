```python
def get_idxlist(xaal):
    frame=xaal
    time=frame.loc[:,('attributes','time_s')]
    shape = np.shape(frame[('attributes','ravi_cropped')][1][5:70,20:85])

    ravi_cropped_array = []
    strain_p1_array = []


    for i in frame.index:
        ravi_cropped_array.append(frame[('attributes','ravi_cropped')][i][5:70,20:85].flatten())
        strain_p1_array.append(frame[('strains','strain_p1')][i][5:70,20:85].flatten())

    df_ravi_cropped = pd.DataFrame(ravi_cropped_array)
    df_strain_p1 = pd.DataFrame(strain_p1_array)

    value_max_ravi_cropped = []
    index_max_list_ravi_cropped = []

    for i in df_ravi_cropped.columns:
        value_max_ravi_cropped.append(df_ravi_cropped[i].max())
        index_max_list_ravi_cropped.append(df_ravi_cropped[i].idxmax())

    value_max_ravi_cropped = np.reshape(value_max_ravi_cropped,(shape))
    index_max_list_ravi_cropped = np.reshape(index_max_list_ravi_cropped,(shape))

    value_max_df_strain_p1 = []
    index_max_list_df_strain_p1 = []

    for i in df_strain_p1.columns:
        value_max_df_strain_p1.append(df_strain_p1[i].max())
        index_max_list_df_strain_p1.append(df_strain_p1[i].idxmax())

    value_max_df_strain_p1 = np.reshape(value_max_df_strain_p1,(shape))
    index_max_list_df_strain_p1 = np.reshape(index_max_list_df_strain_p1,(shape))

    #display(value_max_ravi_cropped, index_max_list_ravi_cropped)
    #display(value_max_df_strain_p1, index_max_list_df_strain_p1)
    return(index_max_list_ravi_cropped,index_max_list_df_strain_p1)
get_idxlist(xaal)
```
